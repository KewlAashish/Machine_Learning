MATPLOTLIB LIBRARY:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [100,200,300]
plt.plot(x,y)
```

```python
housing = pd.DataFrame({"rooms" : [1,1,2,2,2,3,3,3],
                        "prices" : [100,120,190,200,230,310,330,305]})
```

```python
plt.scatter(housing["rooms"] , housing["prices"])
```

```python
plt.xlim(0,2)
plt.ylim(0,200)
plt.plot(x,y , color = "red" , marker = "*" , markersize = 12 , linestyle = "--")
plt.title("Title")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

SEABORN BASICS:

```python
import seaborn as sns
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

```python
sns.distplot(df["age"])
```

```python
plt.figure(figsize = (12,12))
sns.distplot(df["age"] , kde = False , bins= 20 , color = "#344567")
```

```python
sns.countplot(x = "cp" , data = df , hue = "sex" , palette= "terrain")
```

```python
sns.boxplot(x = "target" , y = "thalach" , data = df , hue = "sex")
```

```python
sns.scatterplot(x = "chol" , y = "trestbps" , data = df , hue = "sex")
```

```python
zz = pd.read_csv("iris.csv")
sns.pairplot(zz , hue = "species")
```
